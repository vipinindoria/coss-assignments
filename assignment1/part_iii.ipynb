#### b) Plot the graph of Cache Hit Ratio Vs Cache size with respect to different replacement algorithms. Comment on the graph that is obtained.

```python
a4_dims = (20, 10)
fig, ax = plt.subplots(figsize=a4_dims)
hit_ratio, cache_size = random_table["hit_ratio"].tolist(), random_table["cache_size"].tolist()
fig1 = sns.lineplot(data=random_table, x="hit_ratio", y="cache_size", markers=True)
for i in range(len(hit_ratio)):
    ax.text(hit_ratio[i], cache_size[i], f"{hit_ratio[i], cache_size[i]}")
plt.plot(hit_ratio, cache_size, "-o")
fig1.set(xlabel='Hit Ratio %', ylabel='Cache Size')
plt.title("Replacement Algorithm: Random, Cache Hit Ratio Vs Cache size Graph")
plt.savefig("./output/part_3_section_a_random_plot.png", dpi=300, transparent=True)
```

```python
a4_dims = (20, 10)
fig, ax = plt.subplots(figsize=a4_dims)
hit_ratio, cache_size = fifo_table["hit_ratio"].tolist(), fifo_table["cache_size"].tolist()
fig1 = sns.lineplot(data=fifo_table, x="hit_ratio", y="cache_size", markers=True)
for i in range(len(hit_ratio)):
    ax.text(hit_ratio[i], cache_size[i], f"{hit_ratio[i], cache_size[i]}")
plt.plot(hit_ratio, cache_size, "-o")
fig1.set(xlabel='Hit Ratio %', ylabel='Cache Size')
plt.title("Replacement Algorithm: FIFO, Cache Hit Ratio Vs Cache size Graph")
plt.savefig("./output/part_3_section_a_fifo_plot.png", dpi=300, transparent=True)
```

```python
a4_dims = (20, 10)
fig, ax = plt.subplots(figsize=a4_dims)
hit_ratio, cache_size = lru_table["hit_ratio"].tolist(), lru_table["cache_size"].tolist()
fig1 = sns.lineplot(data=lru_table, x="hit_ratio", y="cache_size", markers=True)
for i in range(len(hit_ratio)):
    ax.text(hit_ratio[i], cache_size[i], f"{hit_ratio[i], cache_size[i]}")
plt.plot(hit_ratio, cache_size, "-o")
fig1.set(xlabel='Hit Ratio %', ylabel='Cache Size')
plt.title("Replacement Algorithm: LRU, Cache Hit Ratio Vs Cache size Graph")
plt.savefig("./output/part_3_section_a_lru_plot.png", dpi=300, transparent=True)
```
